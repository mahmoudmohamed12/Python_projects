7. the number of courses per subject

```python
df['subject'].value_counts()
```

```python
sns.countplot(data=df,x='subject')
plt.xlabel('Subjects')
plt.ylabel('number of courses per subject')
plt.xticks(rotation=60)
plt.show()
```

8. for which level udemy courses are

```python
df['level'].value_counts()
```

```python
sns.countplot(data=df,x='level')
plt.xlabel('levels')
plt.ylabel('number of courses per level')
plt.xticks(rotation=65)
plt.show()
```

9. Display the count of paid and free courses

```python
df['is_paid'].value_counts()
```

```python
sns.countplot(data=df,x='is_paid')
plt.xlabel('is paid or not')
plt.ylabel('number of free and paid courses')
plt.xticks(rotation=60)
plt.show()
```

10. which course has more lecture (free or paid)

```python
df.groupby(['is_paid'])['num_lectures'].mean()
```

11. Which courses have more subscribers(free or paid)

```python
df.groupby('is_paid')['num_subscribers'].count()
```

```python
sns.barplot(data=df,x='is_paid',y='num_subscribers')
plt.show()
```

12. Which level has more subscribers

```python
df.groupby('level')['num_subscribers'].count()
```

```python
sns.barplot(data=df,x='level',y='num_subscribers')
plt.show()
```

13. Find the most popular course title

```python
df.sort_values(by='num_subscribers',ascending=False).iloc[0:1]
```

14. the most 10 popular courses per number of subscribers

```python
df_top_10 = df.sort_values(by='num_subscribers',ascending=False).iloc[0:10]
```

```python
sns.barplot(data=df_top_10,y='course_title',x='num_subscribers')
plt.show()
```

15. Find the course having the most number of reviews

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='subject',y='num_reviews')
plt.show()
```

16. Does the price affect the number of reviews

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='price',y='num_reviews',data=df)
plt.show()
```

17. Find the total number of courses belong to python

```python
len(df[df['course_title'].str.contains('python',case=False)])
```

18. the most 10 popular python courses per number of subscribers

```python
top_10 = df[df['course_title'].str.contains('python',case=False)].sort_values(by='num_subscribers',ascending=False).iloc[0:10,:]
```

```python
sns.barplot(data=top_10,x='num_subscribers',y="course_title")
plt.show()
```
