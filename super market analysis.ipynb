9. Display Daily Sales by Day of the Week

```python
dw_mapping={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    6:'Sun'
    
}
df['day_of_week']=df['Date'].dt.dayofweek.map(dw_mapping)
```

```python
df['day_of_week'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

10. What Will Be the Highest Months for Sales

```python
df['month']=df['Date'].dt.month.map({1:'Jan',2:'Feb',3:'Mar'})
```

```python
df['month'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```
